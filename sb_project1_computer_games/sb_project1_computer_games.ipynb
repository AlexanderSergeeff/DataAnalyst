# Сборный проект №1. Исследование компьютерных игр.

**Описание проекта:**

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Данные до 2016 года.

**Цель проекта:**

Выявить закономерности, определяющие успешность игры. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании магазина "Стримчик" на 2017 год.

**Описание данных:**

`Name` — название игры;

`Platform` — платформа;

`Year_of_Release` — год выпуска;

`Genre` — жанр игры;

`NA_sales` — продажи в Северной Америке (миллионы проданных копий);

`EU_sales` — продажи в Европе (миллионы проданных копий);

`JP_sales` — продажи в Японии (миллионы проданных копий);

`Other_sales` — продажи в других странах (миллионы проданных копий);

`Critic_Score` — оценка критиков (максимум 100);

`User_Score` — оценка пользователей (максимум 10);

`Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board);


*Данные за 2016 год могут быть неполными.*

**Ход исследования:**

- Подготовка данных и изучение общей инфомации;
- Предобработка данных;
- Анализ данных;
- Составление портрета пользователя каждого региона;
- Проверка гипотез;
- Вывод и рекомендации;

Вычислим топ-6 платформ исходя из продаж

```python
#Применим  pivot_table(), отсортируем по убыванию и методом iloc, возьмем срез от 1 до 6.

top6 = df.pivot_table(index='platform', values='total_sales', aggfunc='sum')\
                       .sort_values(by='total_sales', ascending=False).iloc[0:6]
```

```python
#Сбросим индексы

top6_reset_index = top6.reset_index()
```

```python
top6_reset_index
```

```python
#Применим  barplot() из библиотеки seaborn

sns.barplot(data=top6_reset_index, x='platform', y='total_sales')
plt.show()
```

Получим список из уникальных значений топ-6.

```python
#Воспользуемся unique()

top6_unique = top6_reset_index['platform'].unique()
```

Отфильтруем датасет df по этим значениям и запишем в новую переменную.

```python
#Воспользуемся query()

top6_full = df.query('platform in @top6_unique')
```

Построим распределение по годам.

```python
#Создадим сводную и запишем ее в новую переменную

distribution_by_year = top6_full.query('year_of_release != 1985').pivot_table(
    index='year_of_release',
    columns='platform',
    values='total_sales',
    aggfunc='sum'
).sort_values(by='year_of_release', ascending=False)
```

```python
#Применим хитмэп

sns.heatmap(distribution_by_year)
plt.show()
```

**Вывод:** 

- среднее время жизни платформы составляет 11 лет.

## Проверка гипотез

### Cредние пользовательские рейтинги платформ XOne и PC одинаковые

Получим срез по значению платформы "XOne"

```python
#Создадим переменную и отфильтруем по значению XOne 

xone = selected_period[(selected_period['platform'] == 'XOne') & (selected_period['user_score'] > 0)]['user_score']
```

Выведем на экран средний пользовательский рейтинг платформы XOne:

```python
#Применим mean() к переменной

xone.mean()
```

Получим срез по значению платформы "PC"

```python
#Создадим переменную и отфильтруем по значению PC 

pc = selected_period[(selected_period['platform'] == 'PC') & (selected_period['user_score'] > 0)]['user_score']
```

Выведем на экран средний пользовательский рейтинг платформы PC:

```python
#Применим mean() к переменной

pc.mean()
```

Сформулируем нулевую и альтернатиную гипотезы:

- H0: средние пользовательские рейтинги платформ XOne и PC одинаковые.

- H1: средние пользовательские рейтинги платформ XOne и PC разные.

Произведем проверку.

```python
#установим alpha = 5%

alpha = .05

results = st.ttest_ind(xone, pc, equal_var=False)

print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Средние пользовательские рейтинги жанров Action и Sports разные

Получим срез по значению "Action"

```python
#Создадим переменную и отфильтруем по значению Action 

action = selected_period[(selected_period['genre'] == 'Action') & (selected_period['user_score'] > 0)]['user_score']
```

Выведем на экран средний пользовательский рейтинг Action:

```python
#Применим mean() к переменной

action.mean()
```

Получим срез по значению "Sports"

```python
#Создадим переменную и отфильтруем по значению Sports 

sports = selected_period[(selected_period['genre'] == 'Sports') & (selected_period['user_score'] > 0)]['user_score']
```

Выведем на экран средний пользовательский рейтинг Sports:

```python
#Применим mean() к переменной

sports.mean()
```

Сформулируем нулевую и альтернатиную гипотезы:

- H0: средние пользовательские рейтинги платформ Action и Sports одинаковые.

- H1: средние пользовательские рейтинги платформ Action и Sports разные.

Произведем проверку.

```python
#установим alpha = 5%

alpha = .05

results = st.ttest_ind(action, sports, equal_var=False)

print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
